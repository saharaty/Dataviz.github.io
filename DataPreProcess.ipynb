```python
#imports...

import pandas as pd
import matplotlib.pyplot as plt
import os

```

```python
# Set the path to the directory containing your CSV files
csv_directory = '/content/drive/My Drive/DV/archive'

# Initialize an empty list to store DataFrames from each CSV
dataframes = []

# List all CSV files in the directory
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]

# Loop through CSV files, read them, and append to the list
for csv_file in csv_files:
    file_path = os.path.join(csv_directory, csv_file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)

# Now, combined_df contains all data from the 65 CSV files as one table
```
