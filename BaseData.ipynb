```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

###Load data

```python
csv_directory = '/content/drive/MyDrive/ShareFolder/DataViz/archive'
dataframes = []
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]
```

```python
# Loop through CSV files
for csv_file in csv_files:
    file_path = os.path.join(csv_directory, csv_file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Concatenate all DataFrames
df = pd.concat(dataframes, ignore_index=True)
```

```python
#lower case all data
# Assuming df is your DataFrame
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
df
```

```python
# Group by 'common name' and get the first 'scientific name' within each group
mapping = df.groupby('common_name')['scientific_name'].first().to_dict()

# Update 'scientific name' based on the mapping
df['scientific_name'] = df['common_name'].map(mapping)
```
