```python
import pandas as pd
import json

df = pd.read_json("output.json")
```

```python
# Group by city and tree_type
grouped = df.groupby(['city', 'scientific_name']).size().reset_index(name='count')

# top 5 cities by count of trees
top_cities = grouped.groupby('city').sum().nlargest(5, 'count').index

# Filter the data for top 5 cities
top_cities_data = grouped[grouped['city'].isin(top_cities)]

# Get the top 5 trees for each of the top 5 cities
top_trees = top_cities_data.groupby('city').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

top_trees
```
